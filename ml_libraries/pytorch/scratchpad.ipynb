```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
x = torch.rand(4,1)
# x = (x * 100).to(torch.int16)  # scale + convert to int16
y = x * 2

print(x)
print(y)
```

```python
w11 = torch.rand(1,4, requires_grad=True)
b11 = torch.rand(4, requires_grad=True)

w22 = torch.rand(4,1, requires_grad=True)
b22 = torch.rand(1, requires_grad=True)
```

```python
w1 = w11
b1 = b11

w2 = w22
b2 = b22

lr = 0.01     # learning rate
num_epochs = 5
```

```python
for i in range(num_epochs):

    # Forwardpropogation

    h = x.matmul(w1) + b1
    # pass through actiavation function
    # ReLU
    h = torch.relu(h)
    y_pred = h @ w2 + b2

    # Compute loss
    # Mean Squared Error
    loss = torch.mean((y_pred-y)**2)

    loss.backward()  # backpropogation

    # update weights
    with torch.no_grad():
        w1-= lr * w1.grad
        b1 -= lr * b1.grad
        w2 -= lr * w2.grad
        b2 -= lr * b2.grad
    # zero gradients
    w1.grad.zero_()
    b1.grad.zero_()
    w2.grad.zero_()
    b2.grad.zero_()

    print(f"Epoch {i+1}, loss: {loss.item()}")
```

```python
w1 = w11
b1 = b11

w2 = w22
b2 = b22

lr = 0.01     # learning rate
num_epochs = 5
```

```python
params = [w1,b1,w2,b2]

#optimizer
optimizer = optim.SGD(params=params, lr=0.01)
num_epochs = 5
```

```python
for epoch in range(num_epochs):
    # Forwardpropogation

    h = x.matmul(w1) + b1
    # pass through actiavation function
    # ReLU
    h = torch.relu(h)
    y_pred = h @ w2 + b2

    # Compute loss
    # Mean Squared Error
    loss = torch.mean((y_pred-y)**2)

    # zero gradients
    optimizer.zero_grad()

    loss.backward()  # backpropogation

    # update weights
    optimizer.step()

    print(f"Epoch {epoch+1}, loss: {loss.item()}")
```
